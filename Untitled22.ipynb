```python
!pip install CoolProp
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.fftpack import fft, ifft
from scipy.signal import lti, step, impulse
import CoolProp.CoolProp as CP  # Для теплофизических свойств

# Константы
pi = np.pi

class HeatExchangerParams:
    """Класс параметров теплообменника с возможностью изменения нагрузки"""
    def __init__(self, load_percent=100):
        self.load = load_percent / 100  # Коэффициент нагрузки

        # Параметры наружного теплоносителя (первичный контур)
        self.D1 = 120 * self.load  # кг/с
        self.rho1 = 605  # кг/м^3
        self.theta1_in = 259  # °C
        self.alpha1 = 24 * self.load**0.8  # кВт/(м^2∙°C)

        # Параметры внутреннего теплоносителя (вторичный контур)
        self.D2 = 1254 * self.load  # кг/с
        self.rho2 = 749  # кг/м^3
        self.cp2 = 5.3  # кДж/(кг∙°C)
        self.theta2_in = 300 if load_percent == 100 else 282.5  # °C
        self.theta2_out = 265  # °C
        self.alpha2 = 19.9 * self.load**0.8  # кВт/(м^2∙°C)

        # Геометрические параметры трубного пучка
        self.N = 5653  # количество труб
        self.L = 10  # м, длина одной трубы
        self.d_out = 16e-3  # м, наружный диаметр
        self.delta = 1.4e-3  # м, толщина стенки
        self.lambda_metal = 0.016  # кВт/(м∙°C), теплопроводность
        self.c_metal = 0.5  # кДж/(кг∙°C), теплоемкость
        self.rho_metal = 8000  # кг/m^3, плотность

        # Расчет производных параметров
        self._calculate_derived_params()

    def _calculate_derived_params(self):
        """Вычисление производных параметров"""
        self.d_in = self.d_out - 2*self.delta  # внутренний диаметр
        self.S = pi * self.d_in**2 / 4  # площадь сечения потока
        self.A = pi * self.d_in  # внутренний периметр
        self.V = self.S * self.L * self.N  # общий объем

        # Постоянные времени
        self.tau2 = (self.rho2 * self.V) / self.D2  # постоянная времени потока
        self.St2 = (self.alpha2 * self.A * self.L * self.N) / (self.D2 * self.cp2)  # число Стентона
        self.T_metal = (self.rho_metal * self.c_metal * self.delta) / (self.alpha1 + self.alpha2)  # постоянная времени металла

        # Коэффициенты передачи
        self.k1 = self.alpha1 / (self.alpha1 + self.alpha2)
        self.k2 = self.alpha2 / (self.alpha1 + self.alpha2)

        # Коэффициент теплопередачи
        self.K = 1 / (1/self.alpha1 + self.delta/self.lambda_metal + 1/self.alpha2)

def rp_channel_theta_n_to_k(params, omega, z=1):
    """КЧХ РП-модели по каналу θн → θк"""
    j = 1j
    p = j * omega
    exponent = np.exp(-(params.tau2 * p + params.St2) * z)
    return exponent

def rp_channel_theta1s_to_theta2(params, omega, z=1):
    """КЧХ РП-модели по каналу θ1s → θ2"""
    j = 1j
    p = j * omega
    a = params.lambda_metal / (params.rho_metal * params.c_metal)
    r = np.sqrt(p / a)

    ch = np.cosh(r * params.delta)
    sh = np.sinh(r * params.delta)

    W_wall = (params.alpha1 * sh) / (params.alpha1 * sh + params.lambda_metal * r * ch)
    W_flow = np.exp(-(params.tau2 * p + params.St2) * z)

    return W_wall * W_flow * params.k1

def rp_channel_D2_to_theta2(params, omega, z=1):
    """КЧХ РП-модели по каналу D2 → θ2"""
    j = 1j
    p = j * omega
    term = (params.theta2_in - params.theta1_in) * params.St2 / params.D2
    W = term * (1 - np.exp(-(params.tau2 * p + params.St2) * z)) / (params.tau2 * p + params.St2)
    return W

def sp_channel_theta_n_to_k(params, omega):
    """КЧХ СП-модели по каналу θн → θк"""
    j = 1j
    p = j * omega
    return 1 / (params.tau2 * p + 1 + params.St2)

def sp_channel_theta1s_to_theta2(params, omega):
    """КЧХ СП-модели по каналу θ1s → θ2"""
    j = 1j
    p = j * omega
    return params.k1 / ((params.tau2 * p + 1 + params.St2) * (params.T_metal * p + 1))

def sp_channel_D2_to_theta2(params, omega):
    """КЧХ СП-модели по каналу D2 → θ2"""
    j = 1j
    p = j * omega
    term = (params.theta2_in - params.theta1_in) * params.St2 / params.D2
    return term / ((params.tau2 * p + 1 + params.St2) * (params.tau2 * p + 1 + params.St2))

def plot_nyquist(W, title, color='b'):
    """Построение годографа КЧХ"""
    plt.figure(figsize=(8, 8))
    plt.plot(np.real(W), np.imag(W), color=color)
    plt.xlabel('Re(W(jω))')
    plt.ylabel('Im(W(jω))')
    plt.title(title)
    plt.grid(True)
    plt.axis('equal')
    plt.show()

def plot_step_response(params, W_func, model_type, channel_name):
    """Построение переходной характеристики"""
    if model_type == 'RP':
        sys = lti([1], [params.tau2, 1 + params.St2])
    else:
        if channel_name == 'θн → θк':
            sys = lti([1], [params.tau2, 1 + params.St2])
        elif channel_name == 'θ1s → θ2':
            sys = lti([params.k1], [params.tau2*params.T_metal, params.tau2+params.T_metal*(1+params.St2), 1+params.St2])
        elif channel_name == 'D2 → θ2':
            term = (params.theta2_in - params.theta1_in) * params.St2 / params.D2
            sys = lti([term], [params.tau2**2, 2*params.tau2*(1+params.St2), (1+params.St2)**2])

    t = np.linspace(0, 10 * max(params.tau2, params.T_metal), 1000)
    t, y = step(sys, T=t)

    plt.figure(figsize=(10, 6))
    plt.plot(t, y)
    plt.xlabel('Время, с')
    plt.ylabel('Амплитуда')
    plt.title(f'Переходная характеристика {model_type}-модели по каналу {channel_name}\nНагрузка: {params.load*100}%')
    plt.grid(True)
    plt.show()

def static_temperature_distribution(params):
    """Статическое распределение температуры по длине"""
    z = np.linspace(0, 1, 100)
    b = params.K * params.St2 / params.alpha2
    theta2 = params.theta2_in - (params.theta2_in - params.theta1_in) * (1 - np.exp(-b * z)) / (1 - np.exp(-b))
    return z, theta2

def main():
    # Параметры для 100% и 50% нагрузки
    params100 = HeatExchangerParams(100)
    params50 = HeatExchangerParams(50)

    # Диапазон частот для построения КЧХ
    omega = np.logspace(-3, 0.5, 1000)

    # Анализ для 100% нагрузки
    analyze_load_case(params100, "100%")

    # Анализ для 50% нагрузки
    analyze_load_case(params50, "50%")

    # Сравнение статических распределений
    compare_static_distributions(params100, params50)

def analyze_load_case(params, load_str):
    """Анализ для конкретного уровня нагрузки"""
    omega = np.logspace(-3, 0.5, 1000)

    print(f"\nАнализ для нагрузки {load_str}:")

    # 1. РП-модель
    print("\nРП-модель:")
    W_rp_theta = rp_channel_theta_n_to_k(params, omega)
    plot_nyquist(W_rp_theta, f'Годограф КЧХ РП-модели ({load_str} нагрузки)\nпо каналу θн → θк', 'b')

    W_rp_theta1s = rp_channel_theta1s_to_theta2(params, omega)
    plot_nyquist(W_rp_theta1s, f'Годограф КЧХ РП-модели ({load_str} нагрузки)\nпо каналу θ1s → θ2', 'r')

    W_rp_D2 = rp_channel_D2_to_theta2(params, omega)
    plot_nyquist(W_rp_D2, f'Годограф КЧХ РП-модели ({load_str} нагрузки)\nпо каналу D2 → θ2', 'g')

    # 2. СП-модель
    print("\nСП-модель:")
    W_sp_theta = sp_channel_theta_n_to_k(params, omega)
    plot_nyquist(W_sp_theta, f'Годограф КЧХ СП-модели ({load_str} нагрузки)\nпо каналу θн → θк', 'b')

    W_sp_theta1s = sp_channel_theta1s_to_theta2(params, omega)
    plot_nyquist(W_sp_theta1s, f'Годограф КЧХ СП-модели ({load_str} нагрузки)\nпо каналу θ1s → θ2', 'r')

    W_sp_D2 = sp_channel_D2_to_theta2(params, omega)
    plot_nyquist(W_sp_D2, f'Годограф КЧХ СП-модели ({load_str} нагрузки)\nпо каналу D2 → θ2', 'g')

    # 3. Переходные характеристики
    print("\nПереходные характеристики РП-модели:")
    plot_step_response(params, rp_channel_theta_n_to_k, 'RP', 'θн → θк')
    plot_step_response(params, rp_channel_theta1s_to_theta2, 'RP', 'θ1s → θ2')
    plot_step_response(params, rp_channel_D2_to_theta2, 'RP', 'D2 → θ2')

    print("\nПереходные характеристики СП-модели:")
    plot_step_response(params, sp_channel_theta_n_to_k, 'SP', 'θн → θк')
    plot_step_response(params, sp_channel_theta1s_to_theta2, 'SP', 'θ1s → θ2')
    plot_step_response(params, sp_channel_D2_to_theta2, 'SP', 'D2 → θ2')

def compare_static_distributions(params100, params50):
    """Сравнение статических распределений для разных нагрузок"""
    z100, theta2_100 = static_temperature_distribution(params100)
    z50, theta2_50 = static_temperature_distribution(params50)

    plt.figure(figsize=(10, 6))
    plt.plot(z100, theta2_100, 'b-', label='100% нагрузки')
    plt.plot(z50, theta2_50, 'r--', label='50% нагрузки')
    plt.xlabel('Нормированная длина z')
    plt.ylabel('Температура θ2, °C')
    plt.title('Сравнение статических распределений температуры\nвнутреннего теплоносителя для разных нагрузок')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
